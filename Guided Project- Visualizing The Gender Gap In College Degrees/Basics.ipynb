```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
fig = plt.figure(figsize=(12, 24))

cats = [stem_cats, lib_arts_cats, other_cats]

for i,cat in enumerate(cats):
    print(i,cat)
    print(list(range(0,len(cat))))
    for sp in list(range(0,len(cat))):
        ax = fig.add_subplot(6,3,(i+1) + (3* sp))
        ax.plot(women_degrees['Year'], women_degrees[cats[i][sp]], c=cb_dark_blue, label='Women', linewidth=1.5)
        ax.plot(women_degrees['Year'], 100-women_degrees[cats[i][sp]], c=cb_orange, label='Men', linewidth=1.5)
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(cats[i][sp])
        ax.set_yticks([0,100])
        for x in [25,50,75]:
            ax.axhline(x, c=(171/255, 171/255, 171/255), alpha=0.3)
        ax.tick_params(bottom="off", top="off", left="off", right="off")
        if sp == len(cat) - 1:
            ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='on')
        else:
            ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
plt.legend(loc = 'lower right')
fig.savefig('gender_degrees.jpg')
plt.show()
```

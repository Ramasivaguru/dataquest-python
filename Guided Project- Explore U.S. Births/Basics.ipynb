```python
def read_csv(file_name):
    string_list = open(file_name).read().split('\n')
    string_list = string_list[1:]
    final_list = []
    for i in string_list:
        int_fields = []
        string_fields = i.split(',')
        for j in string_fields:
            int_fields.append(int(j))
        final_list.append(int_fields)
    return(final_list)
```

```python
cdc_list = read_csv('US_births_1994-2003_CDC_NCHS.csv')
print(cdc_list[:10])
```

```python
def months_births(list_of_lists):
    births_per_month = {}
    for i in list_of_lists:
        births_per_month[i[1]] = births_per_month.get(i[1],0) + i[4]
    return(births_per_month)
```

```python
cdc_month_births = months_births(cdc_list)
```

```python
def dow_births(list_of_lists):
    day_births = {}
    for i in list_of_lists:
        day_births[i[3]] = day_births.get(i[3],0) + i[4]
    return(day_births)
```

```python
cdc_day_births = dow_births(cdc_list)
print(cdc_day_births)
```

```python
def calc_counts(data, data_column, lookup_column):
    data_for_column = {}
    for i in data:
        data_for_column[i[lookup_column]] = data_for_column.get(i[lookup_column],0) + i[data_column]
    return(data_for_column)
```

```python
cdc_year_births = calc_counts(data = cdc_list,lookup_column = 0,data_column = 4)
```

```python
def calc_min_max(idict):
    min_value = {}
    max_value = {}
    counter = 0
    min_v = 0
    max_v = 0
    for k,v in idict.items():
        if min_v > v or counter == 0:
            min_v = v
            min_value.clear()
            min_value[v] = k
        if max_v < v or counter == 0:
            max_v = v
            max_value.clear()
            max_value[v] = k
        counter += 1
    return(min_value,max_value)
```

```python
   def len_of_cols(list_of_lists):
        l_of_cols = {}
        counter = 0
        no_of_cols = 0
        while no_of_cols < len(list_of_lists[0]):
            l_of_col = 0
            for i,l in enumerate(list_of_lists):
                if len(str(l[no_of_cols])) > l_of_col or counter == 0:
                    counter = 1
                    l_of_col = len(str(l[no_of_cols]))
            l_of_cols[no_of_cols] =  l_of_col
            no_of_cols += 1
        return(l_of_cols)
```

```python
def create_key_value(list_of_lists,keys,value):
    key_value_pair = {}
    l_of_cols = len_of_cols(list_of_lists)
    for i in list_of_lists:
        key_d = ''
        for k in keys:
            l = l_of_cols[k]
            key_d = key_d + str(i[k]).rjust(l,'0')
        key_value_pair[key_d] = i[value]
    return(key_value_pair)
```

```python
create_key_value(ssa_list[:10],(0,1,2,3),4)
```

```python
ssa_key_value_dict = create_key_value(ssa_list,(0,1,2,3),4)
cdc_key_value_dict = create_key_value(cdc_list,(0,1,2,3),4)
```
